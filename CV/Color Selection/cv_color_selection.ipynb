```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
# Read in the image
image = mpimg.imread('C:\\Users\\Rolston Dsouza\\Downloads\\Tensorflow\\self dr(maj prx)\\CV\\parappadi_nitte_1.jpg')
```

```python
#displaying the image
plt.imshow(image)
```

```python
# Grab the x and y size and make a copy of the image
ysize = image.shape[0]
xsize = image.shape[1]
color_select = np.copy(image)
```

```python
print(f"Size of the image: {xsize}*{ysize}")
```

```python
# Define color selection criteria
red_threshold = 200
green_threshold = 200
blue_threshold = 200
```

```python
rgb_threshold = [red_threshold, green_threshold, blue_threshold]
```

```python
# Do a boolean or with the "|" character to identify
# pixels below the thresholds
thresholds = (image[:,:,0] < rgb_threshold[0]) \
            | (image[:,:,1] < rgb_threshold[1]) \
            | (image[:,:,2] < rgb_threshold[2])
color_select[thresholds] = [0,0,0]
```

```python
# Display the image                 
plt.imshow(color_select)
```

```python
mpimg.imsave("parappadi_nitte_1_colorselection.jpg", color_select)
```
